```python
import os
import pickle
import numpy as np

BASE_DIR = "../save/physio/csdi missing=0.1"

metrics = {"RMSE": [], "MAE": [], "CRPS": []}

for fold in range(5):
    fold_dir = os.path.join(BASE_DIR, f"fold{fold}")
    result_file = os.path.join(fold_dir, "result_nsample100.pk")

    if not os.path.exists(result_file):
        raise FileNotFoundError(result_file)

    with open(result_file, "rb") as f:
        rmse, mae, crps = pickle.load(f)

    metrics["RMSE"].append(rmse)
    metrics["MAE"].append(mae)
    metrics["CRPS"].append(crps)

print("\nPer-fold results:")
for i in range(5):
    print(f"Fold {i}: RMSE={metrics['RMSE'][i]:.3f}, "
          f"MAE={metrics['MAE'][i]:.3f}, "
          f"CRPS={metrics['CRPS'][i]:.3f}")

print("\nMean ± SE (rounded to 3 decimals):")
for key in metrics:
    x = np.array(metrics[key], dtype=float)
    mean = x.mean()
    se = x.std(ddof=0) / np.sqrt(len(x))   # ✅ SE
    print(f"{key}: {mean:.3f} ({se:.3f})")


```
